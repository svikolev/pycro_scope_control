```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('model_evals_8_1_23.csv')
df
```

```python

df2 = pd.read_csv('model_evals_soph2ch_8_1_23.csv')
dfc = pd.concat((df,df2),ignore_index = True)
dfc.to_csv('model_evals_1ch2ch_8_2_23.csv')
dfc
```

```python
#desired_model_order = ['Model A', 'Model B', 'Model C']
desired_dataset_order = ['set4_7_10', '5_24', '6_4','6_6']

# Reorder the rows and columns of the DataFrame based on the desired order
#df['model_name'] = pd.Categorical(df['model_name'], categories=desired_model_order, ordered=True)
df['dataset_name'] = pd.Categorical(df['dataset_name'], categories=desired_dataset_order, ordered=True)

# Pivot the DataFrame to convert it into a matrix form with model_name as rows, dataset as columns, and score as values.
pivot_df = df.pivot(index='model_nickname', columns='dataset_name', values='iou')

# Create a heatmap plot
plt.figure(figsize=(10, 6))  # You can adjust the figure size as per your requirement

# Use seaborn's heatmap function to plot the data
sns.heatmap(pivot_df, annot=True, fmt=".2f", cmap='YlGnBu')  # 'annot=True' adds the numbers to the heatmap, 'fmt=".2f"' formats numbers to two decimal places
plt.xlabel('Dataset')  # Add the x-axis label
plt.ylabel('Model')  # Add the y-axis label
plt.title('IOU')  # Add the title for the plot

plt.show()  # Display the plot
```
